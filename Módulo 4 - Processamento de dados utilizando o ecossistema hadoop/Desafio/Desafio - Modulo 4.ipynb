```
df= spark.read.csv("dbfs:/databricks-datasets/COVID/coronavirusdataset/Case.csv", header="true", inferSchema="true")
```

```
df.select('group').distinct().show()
```

```
# Media de casos confirmados
df.agg({'confirmed': 'mean'}).show()
```

```
# Mais casos confirmados por cidade
df.filter(df.group == 'true').groupby('city').count().sort("count",ascending=False).show()
```

```
#média de casos (confirmed), na província (province) de Seoul foi de:
df.filter(df.province == 'Seoul').agg({'confirmed': 'mean'}).show()
```

```
# Carregamento do terceiro dataset - PatientRoute.csv
df3 = spark.read.csv("dbfs:/databricks-datasets/COVID/coronavirusdataset/PatientRoute.csv", header="true", inferSchema="true")
```

```
df3.show()
```

```
#Quantas linhas existem no dataset PatientRoute.csv?
df3.count()
```

```
#No dataset PatientRoute.csv, a maior latitude encontrada é de:
df3.sort(df3.latitude.desc()).show(truncate=False)
```

```
#No dataset PatientRoute.csv, a menor longitude encontrada é de:
df3.sort(df3.longitude.asc()).show(truncate=False)
```

```
# unindo os dois datasets pela coluna de patient_id
dfj = df3.join(df2,df2["patient_id"] == df3["patient_id"])
```

```
dfj.show()
```

```
#Pacientes detectados (type) no aeroporto (airport), e que tiveram caso de infecção (infection_case) como contato com pacientes (contact with patient) retornam uma contagem de:
dfj.filter(dfj.type == 'airport').filter(dfj.infection_case == 'contact with patient').count()
```
